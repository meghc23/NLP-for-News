```python
import unittest, time, random
import urllib.request
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links = []
body = []
imgs = []
final_content = []
website = "https://www.hindustantimes.com/" 

rate = [i/10 for i in range(10)]
cnt = 0
```

```python
from selenium.webdriver.firefox.options import Options

options = Options()
options.binary_location = r"C:\Program Files\Mozilla Firefox\firefox.exe"
browser = webdriver.Firefox(executable_path="geckodriver.exe", options=options)
browser.get("https://www.hindustantimes.com/") # Website subpage link
content = browser.page_source
soup1 = BeautifulSoup(content, "html.parser")
#print(soup1)

#browser.close()
#exit()
```

```python
for main in soup1.findAll('section', attrs = {'class' : 'worldSection sections ht-ad-holder cohort_box'}):
    for block in main.findAll('h3', attrs = {'class' : 'hdg3'}):
    
        name_link = block.find('a', href=True)
        if (name_link != ''):
            body.append((name_link.text).strip())
            links.append(website + (name_link.get('href')).strip())
            
        
df = pd.DataFrame(links, body)
df
```

```python
df.info()
```

```python
cnt = -1
for link in links:
    
    content = ''
    #just for testing --10 links
    cnt = cnt + 1
    if cnt == 37:
        break
    
    print("Fetching link..... : ", body[cnt])
    # time delay before we access the next page..
    time.sleep(random.choice(rate))

    browser.get(link)
    linkcontent = browser.page_source
    soup = BeautifulSoup(linkcontent, "html.parser")
    
    # para 1
    para_1 = soup.find('h2', attrs = {'class' : 'sortDec'})
    if (para_1 != None):
        content = str((para_1.text).strip()) + '' 
    
    # image
    image = soup.find('picture')
    #print(image) --error checker
    if (image != None):
        picture = image.find('img', src=True)
        if (picture != ''):
            imgs.append((picture.get('src')).strip())
        else:
            imgs.append('NaN')
    else:
            imgs.append('NaN')
        
    # remaining para
    for para_rem in soup.findAll('p'):
        content = content + str((para_rem.text).strip()) + '' 
        
    final_content.append(content)
        
# print(final_content)
# print(imgs)
print(final_content)
# final_df = pd.DataFrame(imgs, final_content, links, body)
# final_df
```

```python
final_df = pd.DataFrame(imgs, final_content)
final_df
```
